```python
import seaborn as sns
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
dist = sns.distplot(tips['total_bill'], kde = False, bins = 40)
```

```python
fig = sns.jointplot(x='total_bill' , y='tip', data = tips, kind = 'kde')
```

```python
fig.savefig('Joint Plot.png', dpi = 200)
```

```python
fff = sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

```python
tips.head(2)
```

```python
fff.savefig('Pair Plot.jpg', dpi = 1000)
```
