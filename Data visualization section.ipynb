```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange (0, 100)
y = x*2
z = x**2
```

```python
len(x)
```

```python
%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_title('title')

ax.plot(x,y)
```

```python
fig2 = plt.figure()

ax1 = fig2.add_axes([0,0,1,1])
ax2 = fig2.add_axes([.2,0.5,.2,.2])

ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig3 = plt.figure()

ax_1 = fig3.add_axes([0,0,1,1])
ax_2 = fig3.add_axes([0.2,.5,.4,.4])

ax_1.set_xlabel('X')
ax_1.set_ylabel('Z')
ax_1.plot(x,z)

ax_2.set_xlabel('X')
ax_2.set_ylabel('Y')
ax_2.set_title('zoom')
ax_2.set_xlim([20,22])
ax_2.set_ylim([30,50])
ax_2.plot(x,y)
```

```python
fig,axes = plt.subplots (nrows = 1, ncols = 2)

axes[0].plot(x,y,'b', ls = '--')
axes[1].plot(x,z, 'r', lw = 3)

plt.tight_layout()
```

```python
fig,axes = plt.subplots (nrows = 1, ncols = 2, figsize = (8,1.5))

axes[0].plot(x,y,'b', lw = 3)
axes[1].plot(x,z, 'r', ls = '--')

plt.tight_layout()
```

```python
import numpy as np
x = np.arange (0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('The Plot')
ax.yaxis.set_ticks(np.arange(0,250,50))

ax.plot(x,y)
```

```python
fig2 = plt.figure()

ax1 = fig2.add_axes([0,0,1,1])
ax2 = fig2.add_axes([.2,.5,.2,.2])


```

```python
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.plot(x,y, 'b')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.plot(x,y, 'r')

fig2
```

```python
fig3 = plt.figure()

ax3 = fig3.add_axes([0,0,1,1])
ax4 = fig3.add_axes([.2,.5,.4,.4])

ax3.plot(x,z)
ax3.set_xlabel('x')
ax3.set_ylabel('z')

ax4.plot(x,y)
ax4.set_xlabel('x')
ax4.set_ylabel('y')
ax4.set_title('zoom')
ax4.set_xlim([20,22])
ax4.set_ylim([30,50])
```

```python
fig4,ax = plt.subplots(nrows = 1, ncols = 2)

ax[0].plot(x,y, 'r', lw = 3, ls = '-.')
ax[1].plot(x,z, 'g', lw = 2, ls = '--')

plt.tight_layout()
```

```python
fig4,ax = plt.subplots(nrows = 1, ncols = 2, figsize = ([12,2]))

ax[0].plot(x,y, 'r', lw = 3, ls = '-.')
ax[1].plot(x,z, 'g', lw = 2, ls = '--')

plt.tight_layout()
```
