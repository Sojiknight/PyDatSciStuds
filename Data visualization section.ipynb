```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
```

```python
%matplotlib inline
```

```python
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('A Simple Graph')
plt.plot(x,y)
```

```python
plt.subplot(1,2,1) 
plt.plot(x,y,'y')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
#OBJECT ORIENTED METHOD
```

```python
newplot = plt.figure()
#The figure method takes in four arguments, Left, Bottom, Width, Height that range from 0 to 1
```

```python
axes = newplot.add_axes ([1.0, 1.0, 0.8, 0.8])
```

```python
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('OO-Graph')
```
